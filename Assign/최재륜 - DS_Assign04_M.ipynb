## 2. 첨부한 “videogames.csv” 파일은 여러 게임 플랫폼에서 출시한 게임에 대한 정보를 가지고 있다. 이 파일을 읽어 데이터 프레임 vg에 저장하고 다음에 답하시오.

```python
import seaborn as sns
import matplotlib.pyplot as plt

vg = pd.read_csv('videogames.csv')
```

### (1) 게임의 장르(genre)별 사용자 평가(user_rating)의 평균을 출력하고 장르별로 히스토그램으로 제시하시오. 구간의 개수는 8개로 한다.

```python
print(vg.groupby('genre')['user_rating'].mean())

genre_list = vg['genre'].unique()

for g in genre_list:
    data = vg[vg['genre'] == g]['user_rating']
    
    plt.hist(data, bins=8)
    plt.title(g)
    plt.show()
```

### (2) critic_rating 변수는 사용자로부터 75~100 사이는 긍정적 평가, 50 ~ 74 사이는 보통 평가, 0 ~ 49 사이는 부정적 평가를 받은 것이다. 각 평가를 color를 green, yellow, red로 구분하고 이 기준에 따라 critic_count, user_rating, user_count를 scatter plot으로 제시하시오(교재 p. 245 참조). 

```python
colors = []

for score in vg['critic_rating']:
    if score >= 75:
        colors.append('green')
    elif score >= 50:
        colors.append('yellow')
    else:
        colors.append('red')

# scatter plot
plt.scatter(vg['critic_count'], vg['user_rating'], c=colors)
plt.xlabel('critic_count')
plt.ylabel('user_rating')
plt.title('critic_count vs user_rating')
plt.show()

plt.scatter(vg['critic_count'], vg['user_count'], c=colors)
plt.xlabel('critic_count')
plt.ylabel('user_count')
plt.title('critic_count vs user_count')
plt.show()
```

### (3) 게임의 연령별 등급은 E : 모든 연령, E10+ : 10세이상, T : Teenager, M : 17세 이상, A : 성인을 의미한다. 등급별 사용자 평가(user_rating)을 평균을 출력하고 등급별 히스토그램으로 제시하시오. 구간의 개수는 8개로 한다.

```python
print(vg.groupby('maturity_rating')['user_rating'].mean())

rating_list = vg['maturity_rating'].unique()

for r in rating_list:
    data = vg[vg['maturity_rating'] == r]['user_rating']
    
    plt.hist(data, bins=8)
    plt.title(r)
    plt.show()
```

### (4) 출시 연도(year_of_release)를 제외한 모든 수치 데이터에 대해 다중 피어슨 상관 계수와 scatter matrix과 heat map을 보이고 상관 계수의 절댓값이 가장 큰 2개를 찾아 제시하시오. 

```python
# 숫자 데이터만 선택
num = vg.select_dtypes(include='number')

# year_of_release 제외
num = num.drop('year_of_release', axis=1)

# 상관계수
corr = num.corr()
print(corr)

# scatter matrix
pd.plotting.scatter_matrix(num)
plt.show()

# heatmap
sns.heatmap(corr, annot=True)
plt.show()

# 절댓값 기준으로 정렬
corr_abs = corr.abs()

# 자기 자신 제외
for i in range(len(corr_abs)):
    corr_abs.iloc[i, i] = 0

# 상관 높은 2개 찾기
pairs = corr_abs.unstack().sort_values(ascending=False)

print(pairs.drop_duplicates().head(2))
```
