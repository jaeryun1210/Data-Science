### 1. Import the dataset as a dataframe named df from file.

```python
import pandas as pd

# csv 파일 읽어서 DataFreame으로 저장
df = pd.read_csv('food_consumption.csv')
```

### 2. How many rows and columns are there in the dataframe?

```python
df.shape # df.shape로 (행 개수,열 개수) 형태로 출력된다.
```

### 3. What is the type of data in each column of df?

```python
df.dtypes # 각 컬럼의 데이터 타입 확인
```

### 4. What is the mean co2_emission of the whole dataset?

```python
df['co2_emmission'].mean() # co2_emmission의 평균 계산
```

### 5.How many different kinds of foods are there in the dataset? How many countries are in the dataset?

```python
# 음식 종류 개수와 국가 개수
df[['food_category','country']].nunique() 

#nunique() : 중복 제거 후 고유 값 개수 반환
```

### 6. What is the maximum co2_emmission in the dataset and which food type and country does it belong to?

```python
# co2_emmission이 최대값인 위치 찾기
idx = df['co2_emmission'].idxmax()

# 해당 행 출력
df.loc[idx, ['co2_emmission','food_category','country']]

#idxmax() : 최대값을 가지는 행의 인덱스 반환
#loc[] : 특정 인덱스의 행 데이터 조회
```

### 7. How many countries produce more than 1000 Kg CO2/person/year for at least one food type?

```python
# co2_emmission이 1000 이상인 데이터 필터링
high_co2_countries = df[df['co2_emmission'] > 1000]

# 중복된 국가 제외하고 개수 세기
high_co2_countries['country'].nunique()
```

### 8. Which country consumes the least amount of beef per person per year?

```python
# Beef에 해당하는 데이터만 필터링
beef_data = df[df['food_category'] == 'Beef']

# 쇠고기 소비량이 최소인 국가 찾기
least_beef_country = beef_data.loc[beef_data['consumption'].idxmin(), 'country']

least_beef_country
```

### 9. Which country consumes the most amount of soybeans per person per year?

```python
# Soybeans에 해당하는 데이터만 필터링
soybean_data = df[df['food_category'] == 'Soybeans']

# 콩 소비량이 최대인 국가 찾기
most_soybeans_country = soybean_data.loc[soybean_data['consumption'].idxmax(), 'country']

most_soybeans_country
```

### 10. What is the total emissions of all the meat products (Pork, Poultry, Fish, Lamb & Goat, Beef) in the dataset combined?

```python
# 육류 제품들의 CO2 배출량 합산
meat_categories = ['Pork', 'Poultry', 'Fish', 'Lamb & Goat', 'Beef']

# 해당 음식 범주만 필터링하여 co2_emmission 합계 계산
total_meat_emission = df[df['food_category'].isin(meat_categories)]['co2_emmission'].sum()

total_meat_emission
```

### 11. What is the total emissions of all other (non-meat) products in the dataset combined?

```python
# non-meat인 음식 카테고리 필터링
non_meat_categories = df[~df['food_category'].isin(['Pork', 'Poultry', 'Fish', 'Lamb & Goat', 'Beef'])]

# non-meat 제품들의 CO2 배출량 합산
total_non_meat_emission = non_meat_categories['co2_emmission'].sum()

total_non_meat_emission
```

### 12. What country consumes the most food per person per year (across all food categories)?

```python
# 각 국가별 소비량 합산
total_consumption_per_country = df.groupby('country')['consumption'].sum()

# 가장 많이 소비한 국가 찾기
most_consumed_country = total_consumption_per_country.idxmax()

most_consumed_country
```

### 13. Which food category is the biggest contributor to the above country’s consumption total?

```python
# 가장 많이 소비한 국가에서 해당 국가의 음식 카테고리별 소비량 합산
most_consumed_food_category = df[df['country'] == most_consumed_country].groupby('food_category')['consumption'].sum().idxmax()

most_consumed_food_category
```

### 14. What country produces the most kg C02 per person per year?

```python
# 각 국가별 CO2 배출량 합산
total_co2_per_country = df.groupby('country')['co2_emmission'].sum()

# CO2 배출량이 가장 많은 국가 찾기
most_co2_country = total_co2_per_country.idxmax()

most_co2_country
```

### 15. Which food category is the biggest contributor to the above country’s C02 emissions?

```python
# 위에서 구한 국가에서, 해당 국가의 음식 카테고리별 CO2 배출량 합산
most_co2_food_category = df[df['country'] == most_co2_country].groupby('food_category')['co2_emmission'].sum().idxmax()

most_co2_food_category
```

### 16. What food category produces the most C02 per person per year across all countries?

```python
# 모든 음식 카테고리별 CO2 배출량 합산
total_co2_per_food_category = df.groupby('food_category')['co2_emmission'].sum()

# CO2 배출량이 가장 많은 음식 카테고리 찾기
most_co2_food_category_all = total_co2_per_food_category.idxmax()

most_co2_food_category_all
```

### 17. What food category is consumed the most across all countries per person per year? What food category is consumed the least across all countries?

```python
# 모든 음식 카테고리별 소비량 합산
total_consumption_per_food_category = df.groupby('food_category')['consumption'].sum()

# 가장 많이 소비된 음식 카테고리
most_consumed_food_category = total_consumption_per_food_category.idxmax()

# 가장 적게 소비된 음식 카테고리
least_consumed_food_category = total_consumption_per_food_category.idxmin()

most_consumed_food_category, least_consumed_food_category
```
