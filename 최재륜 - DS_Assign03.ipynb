## 2. 아래 각 문제에 대해 답을 작성하여 Jupyter Notebook 파일로 제출.

### (1) 첨부한 “서울시 대기질 자료 제공_2020-2021.csv” 파일을 읽어 데이터프레임(pm_df)에 저장하시오(단, 일시 열은 날짜시간 형식으로 저장).

```python
import pandas as pd

pm_df = pd.read_csv('서울시 대기질 자료 제공_2020-2021.csv', encoding='cp949', parse_dates=['일시'])
```

### (2) pm_df에서 PM25 컬럼을 제거하시오.

```python
pm_df = pm_df.drop(columns=['초미세먼지(PM25)'])

pm_df
```

### (3) pm_df에서 구분의 값이 ‘성북구’를 제외한 모든 행을 제거하시오.

```python
pm_df = pm_df.loc[pm_df['구분'] == '성북구']

pm_df
```

### (4) pm_df에서 결측치의 개수를 보이시오.

```python
pm_df.isnull().sum()
```

### (5) (3)의 pm_df에서 PM10의 월평균을 구해 데이터프레임 pm10_df에 저장하시오.

```python
pm_df.loc[:,'월'] = pm_df['일시'].dt.month

pm10_df = pm_df.groupby('월')['미세먼지(PM10)'].mean().reset_index()

pm10_df
```

### (6) pm10_df에서 PM10이 가장 높은 달과 낮은 달을 보이시오.

```python
max_month = pm10_df.loc[pm10_df['미세먼지(PM10)'].idxmax()]
min_month = pm10_df.loc[pm10_df['미세먼지(PM10)'].idxmin()]

max_month,min_month
```
